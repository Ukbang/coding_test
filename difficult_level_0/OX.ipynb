```python
def solution(quiz):
    result = []
    for q in quiz:
        question = q.split(' ')
        num1 = int(question[0])
        num2 = int(question[2])
        num3 = int(question[-1])
        if question[1] == '+':
            answer = num1 + num2
        else:
            answer = num1 - num2
        if answer == num3:
            result.append('O')
        else:
            result.append('X')

    return result
```
